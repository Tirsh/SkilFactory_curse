```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value']   
```

```python
exchange_rates('USD')
```

```python
def currency_name(id):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    for item in response.values():
        if item['ID'] == id:
            return item['Name']
  
```

```python
currency_name('R01700J')
```

```python
import requests
from bs4 import BeautifulSoup
def get_actors(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    elem = page.find_all("a", class_="styles_link__1dkjp")
    return [item.text for item in elem]
```

```python
get_actors('https://www.kinopoisk.ru/film/42326/')
```
